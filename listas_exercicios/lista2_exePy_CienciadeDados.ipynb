#### Elabore uma função chamada calcula_desconto que receba dois parâmetros: o valor de um produto e a porcentagem de desconto. A função deve retornar o valor final do produto com o desconto aplicado. Teste a função e imprima o resultado.

```python
# definindo a função para calcular o desconto de um produto 

preco = float(input("Digite o preço do produto: "))
desconto = float(input("Digite a porcentagem de desconto: "))

def calcula_desconto(preco, desconto):
    valor_desconto = preco * (desconto / 100) # Calcula o valor do desconto
    preco_final = preco - valor_desconto # Calcula o preço final após o desconto
    return preco_final

# testando a função

preco_com_desconto = calcula_desconto(preco, desconto)
print(f"O preço final do produto com desconto é: R${preco_com_desconto:.2f}")
```

#### Elabore uma função chamada classifica_nota que receba uma nota de 0 a 10. A função deve retornar "Aprovado" se a nota for maior ou igual a 7, "Recuperação" se estiver entre 5 e 6.9, e "Reprovado" se for menor que 5. Chame a função passando uma nota e imprima o retorno.

```python
# definindo a função para classificação de uma nota de um aluno

def classifica_nota(nota):
    if nota >= 7:
        return "Aprovado"
    elif 6.9 <= nota < 7:
        return "Em Recuperação"
    else:
        return "Reprovado"

# testando a função

nota_aluno = float(input("Digite a nota do aluno: "))
resultado = classifica_nota(nota_aluno)
print(f"O aluno está {resultado}.")
```

#### Elabore um código que receba uma lista de números inteiros mistos (positivos, negativos e zeros). Utilize um laço for com estruturas condicionais if/elif/else para contabilizar quantos números são positivos, quantos são negativos e quantos são exatamente zero. Ao final, imprima esse resumo.

```python
# Pedindo os números ao usuário

entrada = input("Digite números separados por vírgula: ")

# Transformando em lista e convertendo para float

lista_numeros = [float(num) for num in entrada.split(",")]

# Verificando os casos

if all(num > 0 for num in lista_numeros):
    print(f"Todos os números são positivos. Total: {len(lista_numeros)}")

elif all(num < 0 for num in lista_numeros):
    print(f"Todos os números são negativos. Total: {len(lista_numeros)}")

elif all(num == 0 for num in lista_numeros):
    print(f"Todos os números são zeros. Total: {len(lista_numeros)}")

else:
    # Caso seja misto, contamos cada tipo

    positivos = sum(1 for num in lista_numeros if num > 0)
    negativos = sum(1 for num in lista_numeros if num < 0)
    zeros = sum(1 for num in lista_numeros if num == 0)

    print("A lista possui números mistos:")
    print(f"Positivos: {positivos}")
    print(f"Negativos: {negativos}")
    print(f"Zeros: {zeros}")
```
